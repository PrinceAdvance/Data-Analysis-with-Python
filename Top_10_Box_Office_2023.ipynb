```python

import matplotlib.pyplot as plt

# Assuming 'movie_data' is your dataframe and 'budget' & 'worldwide' are numerical columns

# Calculate 'Profit/Loss'
movie_data['movie_profit'] = movie_data['Worldwide'] - movie_data['Budget']

# Sort data by profit in descending order
movie_data_sorted = movie_data.sort_values(by='movie_profit', ascending=False)

# Create the bar chart
plt.figure(figsize=(20, 10))  # Adjust figure size as needed
bar_cont = plt.bar(movie_data_sorted['Release Group'], movie_data_sorted['movie_profit'])

# Annotate bars with profit values
for bar in bar_cont:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 1000, round(yval, 2), ha='center', va='bottom')

# Customize the chart
plt.xlabel('Release Group')
plt.ylabel('movie_profit')
plt.title('movie_profit of Movies (Descending)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Display the chart
plt.show()

```
